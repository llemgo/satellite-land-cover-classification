```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from joblib import dump, load
import seaborn as sns
import shap  # SHAP library
from sklearn.inspection import permutation_importance  # Permutation importance
from matplotlib.colors import ListedColormap

# Load the data
data = pd.read_csv('./All_features_input_sentinel2_train.csv')

# Remove 'class' and 'OBJECTID' and separate features and target
X = data.drop(['class', 'OBJECTID','classid'], axis=1)
y = data['classid']

# Define new experiments for PlanetScope
experiment_1_features = {
    'texture': ['B4_contrast', 'B8_contrast', 'B3_diss', 'B7_var'],
    'indices': ['EVI', 'NDWI', 'BSI'],
    'bands': ['B2', 'B5', 'B8', 'B11'],
    'terrain': ['slope', 'elevation','aspect']
}

experiment_2_features = {
    'texture': ['B4_var', 'B8_diss', 'B3_var', 'B7_contrast'],
    'indices': ['NDVI', 'SAVI', 'NDMI'],
    'bands': ['B3', 'B6', 'B8A', 'B12'],
    'terrain': ['slope', 'elevation', 'aspect']
}

experiment_3_features = {
    'texture': ['B4_diss', 'B8_var', 'B3_contrast', 'B7_diss'],
    'indices': ['EVI', 'NDVI', 'SAVI', 'GNDVI', 'NDMI', 'BSI'],
    'bands': ['B3', 'B4', 'B5', 'B7', 'B8A'],
    'terrain': ['slope', 'elevation', 'aspect']
}

experiment_4_features = {
    'texture': ['B4_contrast', 'B8_contrast', 'B3_var', 'B7_var'],
    'indices': ['EVI', 'NDWI', 'GNDVI'],
    'bands': ['B11', 'B2', 'B5', 'B6', 'B8'],
    'terrain': ['slope', 'elevation', 'aspect']
}


experiments = {
    'Experiment 1': experiment_1_features,
    'Experiment 2': experiment_2_features,
    'Experiment 3': experiment_3_features,
    'Experiment 4': experiment_4_features
}

# Define function to get selected features for each experiment
def get_selected_features(features_dict):
    selected_features = []
    for key in features_dict:
        selected_features.extend(features_dict[key])
    return selected_features

# Loop through each experiment
for exp_name, features_dict in experiments.items():
    print(f"Running {exp_name}")
    selected_features = get_selected_features(features_dict)
    
    # Subset data to selected features
    X_exp = X[selected_features]
    
    # Split the data into training and testing sets (70-30 split)
    X_train, X_test, y_train, y_test = train_test_split(X_exp, y, test_size=0.3, random_state=100, stratify=y)
    
    # Handle missing values by imputation
    imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
    X_train_imputed = imputer.fit_transform(X_train)
    X_test_imputed = imputer.transform(X_test)
    
    # Initialize the Random Forest
    rf = RandomForestClassifier(random_state=100)
    
    # Set up GridSearchCV to find the best parameters, including bootstrap
    param_grid = {
        'n_estimators': [100, 200, 300, 500, 700],
        'max_depth': [None, 10, 20, 30, 40],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'bootstrap': [True],
        'oob_score': [True]  
    }
    
    # Using StratifiedKFold to preserve the percentage of samples for each class
    cv = StratifiedKFold(n_splits=5)
    
    # Setup the grid search
    grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=cv, n_jobs=-1, verbose=2, scoring='accuracy')
    grid_search.fit(X_train_imputed, y_train)
    
    # Best model from grid search
    best_rf = grid_search.best_estimator_
    
    # Evaluate the best model on the training set using cross-validation
    train_cv_scores = cross_val_score(best_rf, X_train_imputed, y_train, cv=cv, scoring='accuracy')
    print("Training Cross-Validation Scores:", train_cv_scores)
    print("Mean Training Cross-Validation Accuracy:", np.mean(train_cv_scores))
    
    # Predict on the test set using the best model
    y_test_pred = best_rf.predict(X_test_imputed)
    
    # Evaluate the model on the test set
    print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
    print("Best Parameters:", grid_search.best_params_)
    print("Classification Report on Test Set:\n", classification_report(y_test, y_test_pred))
    
    # Extract feature importances
    feature_importances = best_rf.feature_importances_
    features = X_exp.columns
    
    # Combine feature names and importance into a DataFrame
    feature_importance_df = pd.DataFrame({
        'Feature': features,
        'Importance': feature_importances
    })
    
    # Sort the DataFrame by importance
    feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
    
    # Print feature importances
    print(feature_importance_df)
    
    # Plot feature importances
    plt.figure(figsize=(10, 7))
    sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.title(f'Feature Importance for {exp_name}')
    plt.show()
    
    # Confusion Matrix for Test Set
    cm = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix on Test Set for {exp_name}')
    plt.show()
    
    # Create a SHAP explainer
    explainer = shap.TreeExplainer(best_rf) 

    # Calculate SHAP values
    shap_values = explainer.shap_values(X_test_imputed)

    # Define custom color palette using Matplotlib's ListedColormap
    custom_cmap = ListedColormap([
        '#1f77b4',  # blue
        '#ff7f0e',  # orange
        '#2ca02c',  # green
        '#d62728',  # red
        '#9467bd',  # purple
        '#8c564b',  # brown
        '#e377c2',  # pink
        '#7f7f7f',  # gray
        '#bcbd22',   # yellow-green
    ])


    # Define class names with respect to their numeric labels directly
    class_names = [
        "Water",          # Class 0
        "Maize fields",   # Class 1
        None,             # Placeholder for Class 2, which is missing
        "Mixed fields",   # Class 3
        "Built up",       # Class 4
        "Shrubland",      # Class 5
        "Trees",          # Class 6
        "Bare",           # Class 7
        "Grass",          # Class 8
        "Rice fields"     # Class 9
    ]

    # If your model skips non-existent classes (like class 2), adjust class_names accordingly
    # Skip 'None' entry if your SHAP values do not expect it
    class_names_adjusted = [name for name in class_names if name is not None]

    # Calculate SHAP values
    explainer = shap.TreeExplainer(best_rf)
    shap_values = explainer.shap_values(X_test_imputed)

    # Plot SHAP values with the adjusted class names
    shap.summary_plot(shap_values, features=X_test_imputed, feature_names=X.columns, class_names=class_names_adjusted,color=custom_cmap)
   # Permutation Importance
    perm_importance = permutation_importance(best_rf, X_test_imputed, y_test, n_repeats=10, random_state=1000, n_jobs=-1)
    
    # Create a DataFrame for Permutation Importance
    perm_importance_df = pd.DataFrame({
        'Feature': features,
        'Importance': perm_importance.importances_mean
    })
    
    # Sort by importance
    perm_importance_df = perm_importance_df.sort_values(by='Importance', ascending=False)
    
    # Print Permutation Importances
    print(perm_importance_df)
    
    # Plot Permutation Importances
    plt.figure(figsize=(10, 7))
    sns.barplot(x='Importance', y='Feature', data=perm_importance_df)
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.title(f'Permutation Importance for {exp_name}')
    plt.show()
```
